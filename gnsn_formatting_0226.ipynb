```python
import csv
with open('orzorz.csv', 'r') as input_file, open('output.csv', 'w', newline='') as output_file:
    reader = csv.reader(input_file)
    writer = csv.writer(output_file)
    for row in reader:
        string = row[0]
        start_index = string.find("""[<Selector xpath=\"descendant-or-self::h1[@class and contains(concat(' ', normalize-space(@class), ' '), ' page-header__title ')]/text()\" data=\'\n\t\t\t\t\t""")
        end_index = string.find("/Voice-Overs\t\t\t\t\'>]")
        extracted_string = string[start_index + len("""[<Selector xpath=\"descendant-or-self::h1[@class and contains(concat(' ', normalize-space(@class), ' '), ' page-header__title ')]/text()\" data=\'\n\t\t\t\t\t"""):end_index]
        # Update the second column of the row with the extracted string
        row[0] = extracted_string
            

        # Write the updated row to the output file
        writer.writerow(row)
```

```python
with open('output.csv', 'r') as input_file, open('output2.csv', 'w', newline='') as output_file:
    reader = csv.reader(input_file)
    writer = csv.writer(output_file)
    for row in reader:
        string = row[1]
        start_index = string.find(",Elemental Skill, ,")
        end_index = string.find("Character Idles,Light Attack, , , , , , , ,Mid Attack, , , ,Heavy Attack, , , ,Climbing, , , , , ,Climbing Breath, , , ,Jumping, , , , , ")
        extracted_string = string[start_index + len(",Elemental Skill, ,"):end_index]
        # Update the second column of the row with the extracted string
        row[1] = extracted_string
        # Write the updated row to the output file
        writer.writerow(row)
```

```python
with open('output.csv', 'r') as input_file, open('output4.csv', 'w', newline='') as output_file:
    reader = csv.reader(input_file)
    writer = csv.writer(output_file)
    # Write header row for output file
    writer.writerow(['name', 'lines', 'E', 'Q', 'sprintStart','sprintEnd', 'glider','treasure', 'lowHP', 'allyLowHP', 'fallen', 'lightHit', 'heavyHit','join'])
    for row in reader:
        # Assume the string you want to extract is in the second column of each row
        string = row[1]
        start_markers = [',Elemental Skill, ,', ',Elemental Burst,', ',Sprint Start,', ',Sprint End, ,',',Deploying Wind Glider, ,',',Opening Treasure Chest, ,',',Low HP, ,',',Ally at Low HP, ,',',Fallen, ,',',Light Hit Taken, ,',',Heavy Hit Taken, ,',',Joining Party, ,']
        end_markers = [',Elemental Burst,', ',Sprint Start,', ',Sprint End, ,',',Deploying Wind Glider, ,',',Opening Treasure Chest, ,',',Low HP, ,',',Ally at Low HP, ,',',Fallen, ,',',Light Hit Taken, ,',',Heavy Hit Taken, ,',',Joining Party, ,',',Character Idles,']
        extracted_strings = []
        for start_marker, end_marker in zip(start_markers, end_markers):
            start_index = string.find(start_marker)
            end_index = string.find(end_marker)
            #print(start_index)
            if start_index != -1 and end_index != -1:
                # Extract the substring between the start and end markers
                extracted_string = string[start_index + len(start_marker):end_index]
                # Add the extracted string to the list of extracted strings
                extracted_strings.append(extracted_string)
            else:
                # If the start or end marker isn't found, add a blank string to the list
                extracted_strings.append('')
        # Create a new row with the original columns and the extracted strings as new columns
        new_row = row + extracted_strings
        # Write the new row to the output file
        writer.writerow(new_row)
```
